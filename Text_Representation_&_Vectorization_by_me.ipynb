### Bag of Words (BOW)

```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({"text":["people watch dswithbappy",
                         "dswithbappy watch dswithbappy",
                         "people write comment",
                          "dswithbappy write comment"],"output":[1,1,0,0]})

df
```

### N-Grams

```python
df = pd.DataFrame({"text":["people watch dswithbappy",
                         "dswithbappy watch dswithbappy",
                         "people write comment",
                          "dswithbappy write comment"],"output":[1,1,0,0]})

df
```

```python
# BI grams
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(ngram_range=(2,2))
```

```python
bow = cv.fit_transform(df['text'])
```

```python
bow.toarray()
```

```python
#vocabulary
print(cv.vocabulary_)
```

```python
#Ti gram
# BI grams
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(ngram_range=(3,3))
```

```python
bow = cv.fit_transform(df['text'])
```

```python
bow.toarray()
```

```python
print(cv.vocabulary_)
```

```python
import os
import re # Import regex for filtering filenames

story = []
# Iterate through files in the /content/ directory where the zip was unzipped
for filename in os.listdir('/content/'):
    # Only process files that match the pattern 'XXXssb.txt'
    if re.match(r'^\d{3}ssb\.txt$', filename):
        filepath = os.path.join('/content/', filename)
        # Use 'with' statement for proper file handling and 'cp1252' encoding
        with open(filepath, 'r', encoding='cp1252') as f:
            corpus = f.read()
            raw_sent = sent_tokenize(corpus)
            for sent in raw_sent:
                story.append(simple_preprocess(sent))
```

```python
story
```

```python
len(story)
```

```python
story[0]
```

```python
model = gensim.models.Word2Vec(
    window=10,
    min_count=2
)
```

```python
model.build_vocab(story)
```

```python
model.train(story, total_examples=model.corpus_count, epochs=model.epochs)
```

```python
vec = model.wv.get_normed_vectors()
```
